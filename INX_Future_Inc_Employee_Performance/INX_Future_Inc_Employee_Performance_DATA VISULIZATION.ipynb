### Importing libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
from matplotlib import rcParams
```

###  Creating a data frame 

```python
pro=pd.read_excel("INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xls")
```

###  Department wise performances

```python
dept=pro.EmpDepartment.unique()
dept
```

```python
dept=pro[pro.EmpDepartment=="Sales"]
dept.head()
```

```python
dept1=dept.iloc[:,[5,-1]]
dept1
```

```python
sns.countplot(x=dept1.PerformanceRating)
```

### Plotting EmpDepartment verses PerformanceRating graph

```python
performance_plot=pd.crosstab(pro.EmpDepartment,pro.PerformanceRating)
```

```python
performance_plot.plot(kind="bar",figsize=(10,10),title="Department wise performances")
plt.savefig("Departmentwiseperformances.png")
```

### Top three Important Factors effecting employee performance

```python

Factor=pro.corr()
```

```python
Factor1=Factor.sort_values("PerformanceRating",ascending=False)
```

```python
Factor2=Factor1.head(4)
Factor2
```

### Plotting graph of performance rating with Top three important factors effecting employee performance

### Factor [EmpEnvironmentSatisfaction]

```python
Factor2_EmpEnvironmentSatisfaction=pd.crosstab(pro.EmpEnvironmentSatisfaction,pro.PerformanceRating)
Factor2_EmpEnvironmentSatisfaction
```

```python
Factor2_EmpEnvironmentSatisfaction.plot(kind="bar",figsize=(10,10)).set_xticklabels("EmpEnvironmentSatisfaction", rotation =60, fontsize='medium')
z=[0,1,2,3]
zq=["low","medium","high","very high"]
plt.xticks(z,zq)
plt.title("EmpEnvironmentSatisfaction verses PerformanceRating graph ")
plt.savefig("EmpEnvironmentSatisfaction verses PerformanceRating graph.png",)
```

### Factor [EmpLastSalaryHikePercent]

```python
Factor2_EmpLastSalaryHikePercent=pd.crosstab(pro.EmpLastSalaryHikePercent,pro.PerformanceRating)
Factor2_EmpLastSalaryHikePercent
```

```python
Factor2_EmpLastSalaryHikePercent.plot(kind="bar",figsize=(10,10))
plt.title("EmpLastSalaryHikePercent verses PerformanceRating graph ")
plt.savefig("EmpLastSalaryHikePercent verses PerformanceRating graph.png")
```

### Factor [EmpWorkLifeBalance]

```python
Factor2_EmpWorkLifeBalance=pd.crosstab(pro.EmpWorkLifeBalance,pro.PerformanceRating)
Factor2_EmpWorkLifeBalance.plot(kind="bar",figsize=(10,10)).set_xticklabels("EmpWorkLifeBalance", rotation =60, fontsize='medium')
x=[0,1,2,3]
xq=["bad","good","better","best"]

plt.xticks(x,xq)
plt.title("EmpWorkLifeBalance verse Performance Rating graph")
plt.savefig("EmpWorkLifeBalance verse Performance Rating graph.png")
```
